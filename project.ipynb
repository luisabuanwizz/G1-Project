```python
while True:
    # asking the user their name, weight and height
    # calculates the bmi
    name = input("Enter your name: ")
    weight = float(input("Enter your weight in kilograms: "))
    height = float(input("Enter your height in meters: "))
    bmi = weight / (height * height)

    # determines if the person is underweight, normal, overweight or obese
    if bmi < 18.5:
      print("Hi!", name)
      print("Your BMI is: ", bmi)
      print("You are underweight!")
    elif bmi <= 24.9:
      print("Hi!", name)
      print("Your BMI is: ", bmi)
      print("You are normal. Nice!")
    elif bmi <= 29.9:
      print("Hi!", name)
      print("Your BMI is: ", bmi)
      print("You are overweight!")
    else:
      print("Hi!", name)
      print("Your BMI is: ", bmi)
      print('You are obese!')

    # asks user if they want to calculate again
    restart = input("Calculate again? (yes/no): ")
    if restart in ["yes", "y", "ye"]:
      continue
    else:
      print("End of calculations, bye!")
      break
```
